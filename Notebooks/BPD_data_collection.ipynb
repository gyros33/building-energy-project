```python
import pandas as pd
import numpy as np
import requests
import json
import os
from config import username, api_key
```

Load in the city data file we found

```python
city_path = os.path.join("Data","usa_cities.csv")
df = pd.read_csv(city_path)
df.head()
```

Find the top five most populous cities in each state, some states on our list didn't have 5 (thanks Manny)

```python
df_test = df.sort_values(by=["province", "pop"],ascending=False)
```

```python
def my_func(pops):
    return list(pops[:5])
df_city_list = df_test.groupby("province").agg({"pop": my_func, "city": my_func})
```

```python
df_city_list = df_city_list.reset_index()
df_city_list
```

Extract a list of these cities

```python
clist = []
for cities in df_city_list["city"]:
    clist.extend(cities)
print(clist)
```

Created the basis for our API calls to the BPD API, and the lists that I will save data to

```python
#Save config information
url = "https://bpd-api.lbl.gov/api/v2/analyze/table"

# Build query URL
# params = {
#     "units":"imperial"
# }

energy_data = pd.DataFrame(clist,columns=["City"])
bad_cities = []
site_eui = []
number_of_people = []
operating_hours = []
occupant_density = []
floor_area = []
source_eui = []
num_buildings = []


headers = {
    "Content-Type":"application/json",
    "Authorization":f"ApiKey {username}:{api_key}",
}

energy_data
```

Required a payload and post to get the data.  Data requires filters and groupings to obtain data.  This code is a test

```python
payload = {
        "filters":{"city":["Austin"]},
        "group_by":["building_class"],
        "analyze_by":"source_eui"
    }
    
response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
res_json = response.json()
res_json
```

```python
res_json["table"][2]
```

Calling the API for the data on commercial buildings, site eui, source eui, occupants, occupants density, floor area, operating hours

```python
for city in clist:
    payload = {
        "filters":{"city":[city]},
        "group_by":["building_class"],
        "analyze_by":"site_eui"
    }
    
    try:
        response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
        res_json = response.json()
        site_eui.append(res_json["table"][2]["mean"])
        print(f"Processing Record : {city}")
        
    except:
        print(f"Missing Data : {city}")
        source_eui.append(np.nan)
        
```

```python
for city in clist:
    payload = {
        "filters":{"city":[city]},
        "group_by":["building_class"],
        "analyze_by":"number_of_people"
    }
    
    try:
        response = requests.post(url=url,data=json.dumps(payload),headers=headers,verify=True)
        res_json = response.json()
        number_of_people.append(res_json["table"][2]["mean"])
        print(f"Processing Record : {city}")
        
    except:
        print(f"Missing Data : {city}")
        number_of_people.append(np.nan)
        
```

```python
energy_data["mean site eui"] = site_eui
energy_data["mean number of people"] = number_of_people
energy_data["mean operating hours"] = operating_hours
energy_data["mean occupant density"] = occupant_density
energy_data["mean floor area"] = floor_area
energy_data["mean source eui"] = source_eui
energy_data
```
