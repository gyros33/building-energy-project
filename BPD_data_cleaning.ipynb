```python
import pandas as pd
import numpy as np
import requests
import json
import os
from config import username, api_key
from city_list import slist, clist, alist
```

Load up our BPD data for cleaning

```python
city_path = os.path.join("Data","bpd_data.csv")
df = pd.read_csv(city_path)
df
```

Add this list so we can combine the BPD Data with climate data AKA merge on this field

```python
df["state"]=alist
df["name"]=df["City"]+", "+df["state"]+" US"
df
```

Grab the weather data

```python
weather_path = os.path.join("Data","weather_data.csv")
df2 = pd.read_csv(weather_path)
df2 = df2.drop("id",axis=1)
#df2.rename({"name":lookup})
df2
```

Load up the residential data

```python
res_path = os.path.join("Data","bpd_data_residential.csv")
dfres = pd.read_csv(res_path)
dfres["City"] = df["name"]
dfres = dfres.rename(columns={"City":"name",
                             "mean source eui":"mean source eui res",
                             "mean site eui":"mean site eui res"})
dfres
```

Load up the additional data for further analysis

```python
ad_path = os.path.join("Data","bpd_data_additional.csv")
dfad = pd.read_csv(ad_path)
dfad = dfad.rename(columns={"city":"City"})
dfad
```

Merge!!

```python
df3 = pd.merge(df, df2, on="name",how="inner")
df3 = pd.merge(df3, dfres, on="name",how="inner")
df3 = pd.merge(df3, dfad, on="City",how="inner")
df3 = df3.drop("name",axis=1)
df3 = df3.rename(columns={"mean electirc eui":"mean electric eui"})
df3
```

Were going to take out some columns we don't need anymore and rearrage them

```python
df4 = df3[['City', 'state', 'mean site eui', 'mean site eui res', 'mean source eui','mean source eui res', 'mean fuel eui','mean fuel eui res', 'mean electric eui','mean electric eui res', 'CDD', 'HDD', 'Tavg', 'Trange', 'Tmax', 'Tmin', 'Prcp', 'Snow']]
df4 = df4.rename(columns={"City":"city"})
```

Doing some cleaning, we can't make comparisons if the data in site eui doesn't exist
df5 is for commercial
df6 is for residential

```python
df5 = df4.dropna(axis=0,subset=["mean site eui"])
df6 = df4.dropna(axis=0,subset=["mean site eui res"])
df5
```

Save files

```python
df5.to_csv(os.path.join("Data","combined_data.csv"),index=False)
```

```python
df6.to_csv(os.path.join("Data","combined_data_residential.csv"),index=False)
```
